```python
import pandas as pd
data = {
    'name' : ['sakthivel','nithi','sam'],
    'age' : [19,19,19],
    'city': ['bangalore','bangalore','madurai']
}
df = pd.DataFrame(data)
df['salary'] = [50000,100000,600000]
df
```

```python
df.drop('salary',axis = 1)
```

```python
df  # See even though we drop the salary column ,we can get the salary while print 
```

```python
#so if we want to drop it permanently we need to add 
df.drop('salary',axis = 1,inplace=True)
df
```

```python
df = df.rename(columns = {'name':'student_name'})  #change the columns name
df
```

```python
##change the data types
df.dtypes
```

```python
df['age'] = df['age'].fillna(df['age'].mean()).astype(float)  ## here we have first fill the missing values and then we do the typecast of the columns 
```

```python
df
```

```python
df['new_age'] = df['age'].apply(lambda x:x**2)  ## in this apply we can as the function , as i have passed the lambda function as of now 
df
```

```python
def cube(*args):
    return [i**2 for i in args]
df['new_age1'] = df['age'].apply(cube)    ## in this case we can give inbuilt funtion , custom function...
df
```

```python
df.groupby('student_name')['age'].mean() ## This is group by student_name
```

```python
df.groupby(['student_name','city'])['age'].mean() ##This is group by student_name,city
```

```python
df.groupby(['student_name'])['age'].agg(['mean','sum','count'])
```

```python
##Merge and join the dataframe
df1 = pd.DataFrame({'key':['a','b','c'],'values1':[1,2,3]})
df2 = pd.DataFrame({'key':['a','b','d'],'values2':[4,5,6]})
pd.merge(df1,df2,on = "key",how = "inner")
```

```python
pd.merge(df1,df2,on = "key",how = "outer")
```

```python
pd.merge(df1,df2,on = "key",how = "left")  ## the left dataframe given more priority
```

```python
pd.merge(df1,df2,on = "key",how = "right") ## the right dataframe given more prirority
```
