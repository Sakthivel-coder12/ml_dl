```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('sales_data.csv')
df.head()
```

```python
grp = df.groupby('Product Category')['Total Revenue'].sum()
grp
```

```python
grp.plot(kind = "bar",color = "lightskyblue",edgecolor = "black")
```

```python
grp_by_date = df.groupby('Date')['Total Revenue'].sum().reset_index()  # this reset index makes the data into two columns (beautifully)
grp_by_date
```

```python
grp_by_date.plot(kind = "hist",color = "orange",edgecolor= "black")
```

```python
## create scatter plot
import matplotlib.pyplot as plt 
import seaborn as sns
df = sns.load_dataset('tips')
sns.scatterplot(x='total_bill',y = 'tip',data = df)
plt.show()
```

```python
#line plot
sns.lineplot(x = 'size',y= 'total_bill',data = df)
```

```python
##catogorical plot
##barplot

sns.barplot(x = 'day',y = 'total_bill',data = df)
plt.title("The days VS total_bill")
plt.xlabel('days')
plt.ylabel('total_bill')
plt.grid(which='minor')
plt.minorticks_on()
```

```python
sns.boxplot(x= 'day',y = 'total_bill',data = df)  
```

```python
sns.boxenplot(x= 'day',y = 'total_bill',data = df)
```

```python
sns.violinplot(x= 'day',y = 'total_bill',data = df)
```

```python
sns.histplot(df['total_bill'],bins=10,kde= True)    # the reason for getting the curve is "kde"
```

```python
sns.kdeplot(df['total_bill'],fill=True,color='red',edgecolor='black')
plt.title('The total_bills')
```

```python
sns.kdeplot(data=df, x='total_bill', fill=True,bw_adjust=0.5,color='red',alpha = 0.9, linewidth=2, label='Total Bill',edgecolor='blue')
```

```python
sns.pairplot(df,hue = 'smoker',kind="scatter",diag_kind='kde',corner=False)    ## In the hue you have to give the any column in the data set 
##diag_kind	Kind of plot on the diagonal: 'auto', 'hist', or 'kde'.
##corner	If True, only shows lower triangle of plots (less cluttered).
'''hue_order	Order of categories in the hue variable.
plot_kws	Dictionary of keyword arguments for the off-diagonal plots.
diag_kws	Dictionary of keyword arguments for the diagonal plots.
markers	Marker styles for different levels of the hue variable.
height	Size (in inches) of each facet.
aspect	Aspect ratio of each facet.'''
```

```python
sns.pairplot(df) 
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df = pd.read_csv('sales_data.csv')
plt.figure(figsize = [10,6])
sns.barplot(x='Product Category', y='Total Revenue', data=df, color='red')
plt.title('Product Category Vs Total Revenue')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.show()

'''sns.barplot(x='Product Category', y='Total Revenue', data=df, palette='Set2')

Some popular Seaborn palettes:

    'pastel'

    'deep'

    'muted'

    'bright'

    'Set1', 'Set2', 'Set3'

    'husl', 'coolwarm', 'Blues', 'Greens', 'Reds''''
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=[10, 6])
sns.barplot(x='Product Category', y='Total Revenue', data=df,
            hue='Product Category', palette='pastel', legend=False)
plt.title("Total Revenue by Product Category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```
